#**Forecasting Website Traffic**
Forecasting traffic on a website during a time period helps a business prepare marketing strategies in advance to increase traffic on their website when the expected traffic is low.

Here is a daily traffic data of a website we collected. Below are all the features in the data:

1. `Date`: Date of the record
2. `Views`: Total number of views in the day

Forecast traffic on the website by analyzing the past year's traffic data of the website.

```python
import pandas as pd
import matplotlib.pyplot as plt
import plotly.express as px
import plotly.graph_objects as go
from statsmodels.tsa.seasonal import seasonal_decompose
from statsmodels.graphics.tsaplots import plot_pacf
from statsmodels.tsa.arima_model import ARIMA
import statsmodels.api as sm
```

```python
data = pd.read_csv("dataset.csv")
print(data.head())
```

```python
data["Date"] = pd.to_datetime(data["Date"],
                              format="%d/%m/%Y")
print(data.info())
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15, 10))
plt.plot(data["Date"], data["Views"])
plt.title("Daily Traffic of dataset.com")
plt.show()
```

```python
result = seasonal_decompose(data["Views"],
                            model='multiplicative',
                            period = 30)
fig = plt.figure()
fig = result.plot()
fig.set_size_inches(15, 10)
```

```python
pd.plotting.autocorrelation_plot(data["Views"])
```

```python
plot_pacf(data["Views"], lags = 100)
```

```python
p, d, q = 5, 1, 2
model=sm.tsa.statespace.SARIMAX(data['Views'],
                                order=(p, d, q),
                                seasonal_order=(p, d, q, 12))
model=model.fit()
print(model.summary())
```

```python
predictions = model.predict(len(data), len(data) + 50)
print(predictions)
```

```python
data["Views"].plot(legend = True, label = "Training Data",
                   figsize = (15, 10))
predictions.plot(legend = True, label = "Predictions")
```
